# Implementing market basket analysis

```python
#Loading neccesary packages
import numpy as np
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

# BASKET ANALYSIS >199

```python
order_df = pd.read_csv('order_processed.csv')
order_df.head()
```

```python
order_df.info()
```

```python
user_df = pd.read_csv('user_processed.csv')
user_df.head()
```

```python
listen_df = pd.read_csv('listening_processed.csv')
listen_df.head()
```

```python
order_df1 = order_df.groupby(['UserID (FK)'], as_index=False).sum()
order_df1.head()
```

```python
order_df2 = order_df1.loc[order_df1['Total Amount'] > 199000]
order_df2.head()
```

```python
playlist_df = listen_df.merge(order_df2,how='inner',on='UserID (FK)',)
playlist_df.info()
```

```python
abc = playlist_df[['UserID (FK)', 'Playlist Name']]
abc.head()
```

```python
b = abc.values.tolist()
```

```python
from collections import defaultdict
d = defaultdict(list)
for k, v in b:
    d[k].append(v)
for k in d:
  d[k] = list(set(d[k]))
```

```python
df1 = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in d.items() ]))
df1 = df1.transpose()
```

```python
d.values()
```

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
onehot = te.fit_transform(d.values())
onehot = pd.DataFrame(onehot, columns = te.columns_)
onehot
```

```python
frequent_itemsets=apriori(onehot, min_support=0.01, use_colnames=True, max_len=3)
frequent_itemsets
```

```python
#Generating Association rule with mlxtend's association_rules
rules=association_rules(frequent_itemsets)
# Display 25 rules with highest lift
rules.sort_values(by=['lift'], ascending=False).head(25)
```

```python
rules['antecedent'] = rules['antecedents'].apply(lambda antecedent: list(antecedent))
rules['consequent'] = rules['consequents'].apply(lambda consequent: list(consequent))
coords1 = rules[['antecedent','consequent']]
coords1.head(25)
```

```python
from google.colab import files
coords1.to_csv('basketanalysis1.csv')
files.download('basketanalysis1.csv')
```
