```python
from collections import defaultdict
d = defaultdict(list)
for k, v in df2:
    d[k].append(v)
for k in d:
  d[k] = list(set(d[k]))
```

```python
df2 = pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in d.items() ]))
df2 = df2.transpose()
df2
```

```python
from mlxtend.preproces import TransactionEncoder

te = TransactionEncoder()
onehot = te.fit_transform(d.values())
onehot = pd.DataFrame(onehot, columns = te.columns_)
```

```python
onehot
```

```python
frequent_itemsets=apriori(onehot, min_support=0.005 use_colnames=True, max_len=3)
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets)
rules
```

```python
rules['antecedent'] = rules['antecedents'].apply(lambda antecedent: list(antecedent))
rules['consequent'] = rules['consequents'].apply(lambda consequent: list(consequent))
coords1 = rules[['antecedent','consequent']]
```
